```python
# -*- coding: utf-8 -*-
"""
Created on Mon Dec 30 11:10:16 2019

@author: Sarfo Senior
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_file = pd.read_csv('georgia.csv') #Import data from georgia.csv in the same directory
input_list = ['temperature','pH']
output_list = ['dissolved_oxygen']

#Extract parameters from data_file
input_values = np.array(data_file[input_list]) 
output_values = np.array(data_file[output_list])

plt.plot(input_values, output_values)
plt.legend(["temperature", "pH"], loc=4)
plt.xlabel("features")
plt.ylabel("dissovled oxygen")
plt.show()
   
```
